```python
import numpy as np
```

```python
# Load the CSV file into a numpy array, ignoring the header row
data = np.loadtxt('revenue_expenses.csv', delimiter=',', skiprows=1)
```

```python
total_revenue = np.sum(data[:, 0])
total_expenses = np.sum(data[:, 1])
```

```python
net_income = total_revenue - total_expenses
```

```python
avg_revenue = np.mean(data[:, 0])
avg_expenses = np.mean(data[:, 1])
```

```python
std_revenue = np.std(data[:, 0])
std_expenses = np.std(data[:, 1])
```

```python
min_revenue = np.min(data[:, 0])
max_revenue = np.max(data[:, 0])
min_expenses = np.min(data[:, 1])
max_expenses = np.max(data[:, 1])
```

```python
# Print the results
print("Total Revenue: ${:.2f}".format(total_revenue))
print("Total Expenses: ${:.2f}".format(total_expenses))
print("Net Income: ${:.2f}".format(net_income))
print("Average Monthly Revenue: ${:.2f}".format(avg_revenue))
print("Average Monthly Expenses: ${:.2f}".format(avg_expenses))
print("Standard Deviation of Monthly Revenue: ${:.2f}".format(std_revenue))
print("Standard Deviation of Monthly Expenses: ${:.2f}".format(std_expenses))
print("Minimum Monthly Revenue: ${:.2f}".format(min_revenue))
print("Maximum Monthly Revenue: ${:.2f}".format(max_revenue))
print("Minimum Monthly Expenses: ${:.2f}".format(min_expenses))
print("Maximum Monthly Expenses: ${:.2f}".format(max_expenses))
```

```python

import matplotlib.pyplot as plt

# Load data from CSV file
data = np.genfromtxt('revenue_expenses_data_1.csv', delimiter=',', skip_header=1)

# Extract the revenue and expense data
revenue = data[:,0]
expenses = data[:,1]

# Create a line chart of revenue and expenses over time
months = np.arange(1, 13)
plt.plot(months, revenue, label='Revenue')
plt.plot(months, expenses, label='Expenses')

# Set the chart title and axis labels
plt.title('Monthly Revenue and Expenses')
plt.xlabel('Month')
plt.ylabel('Amount')

# Add a legend to the chart
plt.legend()

# Display the chart
plt.show()
```

```python
#Graph Data Comparison
# Load data from csv file
data = np.loadtxt('revenue_expenses_data_1.csv', delimiter=',', skiprows=1, usecols=(1, 2), dtype=float)

# Separate data into revenue and expenses
revenue = data[:, 0]
expenses = data[:, 1]

# Set up bar graph
bar_width = 0.35
index = np.arange(len(revenue))

# Plot revenue and expenses on same graph
plt.bar(index, revenue, bar_width, label='Revenue')
plt.bar(index + bar_width, expenses, bar_width, label='Expenses')

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Monthly Revenue and Expenses')

# Add legend and grid
plt.legend()
plt.grid(True)

# Display graph
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from CSV files
revenue_data = np.loadtxt('revenue_expenses.csv', delimiter=',', skiprows=1, usecols=(1,))
expenses_data = np.loadtxt('revenue_expenses_data_1.csv', delimiter=',', skiprows=1, usecols=(1,))

# Calculate total revenue and expenses
total_revenue = np.sum(revenue_data)
total_expenses = np.sum(expenses_data)

# Calculate revenue and expenses as percentages of total
revenue_percent = (total_revenue / (total_revenue + total_expenses)) * 100
expenses_percent = (total_expenses / (total_revenue + total_expenses)) * 100

# Create a pie chart
labels = ['Revenue', 'Expenses']
sizes = [revenue_percent, expenses_percent]
colors = ['#ff9999','#66b3ff']
explode = (0.05, 0) 

fig1, ax1 = plt.subplots()
ax1.pie(sizes, colors = colors, explode=explode, labels=labels, autopct='%1.1f%%', startangle=90)
ax1.axis('equal')
plt.title('Revenue vs Expenses')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from CSV files
data1 = np.loadtxt('revenue_expenses.csv', delimiter=',', skiprows=1)
data2 = np.loadtxt('revenue_expenses_data_1.csv', delimiter=',', skiprows=1)

# Extract data columns
months = data1[:, 0]
revenue1 = data1[:, 1]
expenses1 = data1[:, 2]
revenue2 = data2[:, 1]
expenses2 = data2[:, 2]

# Set bar width
bar_width = 0.35

# Set index values for x-axis
index = np.arange(len(months))

# Plot the bar graphs
plt.bar(index, revenue1, bar_width, label='Revenue 1')
plt.bar(index + bar_width, revenue2, bar_width, label='Revenue 2')
plt.bar(index, -expenses1, bar_width, label='Expenses 1')
plt.bar(index + bar_width, -expenses2, bar_width, label='Expenses 2')

# Add labels, title and legend
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Monthly Revenue and Expenses')
plt.xticks(index + bar_width / 2, months)
plt.legend()

# Display the plot
plt.show()
```
