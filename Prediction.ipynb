Revenue Prediction

```python
import numpy as np
from sklearn.model_selection import train_test_split

# Load revenue data
revenue_data = np.loadtxt('revenue_expenses.csv', delimiter=',', skiprows=1)


# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(revenue_data[:, 0], revenue_data[:, 1], test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

# Train linear regression model
regressor = LinearRegression()
regressor.fit(X_train.reshape(-1, 1), y_train.reshape(-1, 1))

```

```python
# Make predictions on testing data
y_pred = regressor.predict(X_test.reshape(-1, 1))
```

```python
import matplotlib.pyplot as plt

# Plot predicted revenue and actual revenue
plt.plot(X_test, y_pred, color='red', label='Predicted Revenue')
plt.scatter(X_test, y_test, color='blue', label='Actual Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.title('Monthly Revenue Prediction')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Load revenue data
revenue_data = np.loadtxt('revenue_expenses.csv', delimiter=',', skiprows=1)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(revenue_data[:, 0], revenue_data[:, 1], test_size=0.2, random_state=0)

# Train the linear regression model
model = LinearRegression()
model.fit(X_train.reshape(-1, 1), y_train.reshape(-1, 1))

# Predict the revenue data using the linear regression model
y_pred_train = model.predict(X_train.reshape(-1, 1))
y_pred_test = model.predict(X_test.reshape(-1, 1))

# Plot the training data
plt.plot(X_train, y_train, 'o', label='Revenue Data')

# Plot the predicted revenue data
plt.plot(X_train, y_pred_train, label='Predicted Revenue Data')

# Set the title and axis labels
plt.title('Monthly Revenue Prediction')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.legend()

# Show the plot
plt.show()
```
